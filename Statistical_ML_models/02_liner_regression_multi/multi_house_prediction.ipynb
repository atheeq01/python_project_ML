### import the degeneracies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

### read CSV

```python
df = pd.read_csv("assets/homeprices.csv")
```

```python
df.head
```

#### fill the missing value

```python
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
```

```python
df.head()
```

### draw the scatter plot

```python
sns.pairplot(data=df,
             kind="reg",
             diag_kind='kde',
             plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha': 0.5}},
             corner=True
             )
```

```python
liner_lm = LinearRegression()
```

```python
liner_lm.fit(df[['area','bedrooms','age']],df['price'])
```

```python
liner_lm.predict(pd.DataFrame([[3000,3,40],[2500,4,5]],columns=['area','bedrooms','age']))
```
