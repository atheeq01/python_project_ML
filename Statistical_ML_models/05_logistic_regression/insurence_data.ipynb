### import dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

### read teh csv

```python
df = pd.read_csv("assets/insurance_data.csv")
df.head()
```

```python
sns.lmplot(
    data=df,
    x="age",
    y="bought_insurance",
    scatter_kws={"color":"blue","alpha":0.5},
    line_kws={"color":"red","linestyle":"--","linewidth":2,"alpha":.5}
)
```

### split the data

```python
X_train, X_test, y_train, y_test = train_test_split(df[["age"]] , df.bought_insurance,test_size=0.2,random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
X_test
```

### train the data

```python
l_reg = LogisticRegression()
```

```python
l_reg.fit(X_train,y_train)
```

```python
l_reg.predict(X_test)
```

```python
l_reg.score(X_test,y_test)
```

#### model.coef_ indicates value of m in y=m*x + b equation

```python
l_reg.coef_
```

#### model.intercept_ indicates value of b in y=m*x + b equation

```python
l_reg.intercept_
```

#### sigmoid

```python
import math
def sigmoid(x):
     return 1/(1+math.exp(-x))
```

```python
def prediction_function(age):
    z = 0.113077 * age -4.15311919 # 0.04150133 ~ 0.042 and -1.52726963 ~ -1.53
    y = sigmoid(z)
    return y
```
