### import dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

### read the csv

```python
df= pd.read_csv("assets/houseprice.csv")
```

```python
df.head()
```

```python
plt.scatter(df["area"], df["price"])
plt.xlabel("Area")
plt.ylabel("Price")
```

```python
sns.jointplot(
    x="area",
    y="price",
    data=df,
    alpha=0.5,
)
```

```python
sns.lmplot(
    x="area",
    y="price",
    data=df,
    scatter_kws={"color": "blue", "alpha": 0.5},
    line_kws={"color": "red", "alpha": 1,"linewidth": 2,"linestyle": "--"}
)
```

```python
print("df[['area']]:",df[["area"]].shape,"--\t",type(df[["area"]]))
print("df.area:",df.area.shape,"--\t",type(df.area))
print("df['area']:",df["area"].shape,"--\t",type(df["area"]))
```

```python
liner_lm = LinearRegression()
```

```python
liner_lm.fit(df[["area"]], df.price)
```

```python
d = pd.DataFrame({"area":[3300,1845]})
```

```python
liner_lm.predict(d)
```

### predict the value from csv

```python
d = pd.read_csv("assets/test_areas.csv")
```

```python
p = liner_lm.predict(d)
```

```python
d["price"] = np.round(p,2)
```

### save the csv

```python
d.to_csv("assets/predict_areas.csv",index=False)
```
