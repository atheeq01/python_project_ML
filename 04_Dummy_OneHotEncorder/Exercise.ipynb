### import dependencies

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
```

### read csv

```python
df = pd.read_csv("assets/carprices.csv", names=["model","mileage","price","age"],header=0)
```

```python
df.head(8)
```

```python
df.model.unique()
```

### label the text to numaric

```python
labelEncoder = LabelEncoder()
```

```python
df.model = labelEncoder.fit_transform(df.model)
```

```python
df.model.unique()
```

```python
X = df.drop(["price"],axis=1)
y = df.price
```

```python
ohe = OneHotEncoder()
```

```python
ct = ColumnTransformer([("model",ohe,[0])],remainder="passthrough")
```

```python
X = ct.fit_transform(X)
```

```python
X= X[:,1:]
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
X
```

<b>Price of mercedez benz that is 4 yr old with mileage 45000</b>

```python
lr.predict([[0,1,45000,4]])
```

<b>Price of BMW X5 that is 7 yr old with mileage 86000</b>

```python
lr.predict([[1,0,86000,7]])
```
