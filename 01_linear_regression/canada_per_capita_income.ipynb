### import the dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from asttokens import LineNumbers
from sklearn.linear_model import LinearRegression

from linear_regression.House_prediction import liner_lm
```

### read the csv

```python
df = pd.read_csv("assets/canada_per_capita_income.csv",names=["year","income"],header=0)
```

```python
df.head()
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("income")
plt.scatter(df["year"], df["income"])
```

```python
sns.lmplot(
    x="year",
    y="income",
    data=df,
    scatter_kws={"color": "blue","alpha":0.5},
    line_kws={"color":"red","linewidth":2,"linestyle":"--","alpha":0.5},
)
```

```python
liner_lm = LinearRegression()
```

```python
liner_lm.fit(df[["year"]],df.income)
```

```python
d = pd.DataFrame({"year":[2015,2020,2021]})
liner_lm.predict(d)
```
